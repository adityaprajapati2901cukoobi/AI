```python
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import association_rules,apriori

data=pd.read_excel('data/retail.xlsx')
data.head()
```

```python
basket=(data[data['Country']=='United Kingdom']
        .groupby(['InvoiceNo','Description'])['Quantity']
        .sum().unstack().reset_index().fillna(0)
        .set_index('InvoiceNo'))
basket
```

```python
def hot_encode(x):
    if(x<=0):
        return 0
    if(x>=0):
        return 1
    
basket_encoded=basket.applymap(hot_encode)
basket_UK=basket_encoded
basket_UK
```

```python
freq_item = apriori(basket_encoded,min_support=0.07,use_colnames=True)
rules=association_rules(freq_item,metric='confidence',min_threshold=1)
rules.head()
```
