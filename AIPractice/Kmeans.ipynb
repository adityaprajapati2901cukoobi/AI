```python
import pandas as pd
data=pd.read_csv('data/OnlineBusinessCustomers.csv',usecols=['Genre','Age','Income','Spends'])
data.head()
```

```python
import seaborn as sns
sns.scatterplot(data=data,x='Age',y='Income',hue='Spends')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
from sklearn import preprocessing

X_train,X_test,y_train,y_test=train_test_split(data[['Spends','Income']],data[['Age']],test_size=0.33,random_state=1)
X_train_norm=preprocessing.normalize(X_train)
X_test_norm=preprocessing.normalize(y_train)
kmeans=KMeans(n_clusters=2,random_state=0,n_init='auto')
kmeans.fit(X_train_norm)
```

```python
sns.scatterplot(data=X_train,x='Income',y='Spends',hue=kmeans.labels_)

```

```python
#hierarical Clustering
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

data_scalar=StandardScaler()
new=data_scalar.fit_transform(X_train_norm)

linkage_data=linkage(new,metric='euclidean',method='single')
dendrogram(linkage_data)
plt.show()
```
